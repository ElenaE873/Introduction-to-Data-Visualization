##### Advantages of Seaborn

- works well with pandas data structures
- built on top of matplotlib

##### Important

- seaborn is not as good if data is not "tidy"
- therefore data should be cleaned up before applying seaborn

##### seaborn with a lists of values

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#building a scatterplot

sns.scatterplot(x=variable1, y=variable2)
plt.show()
#good for comparing to types of variables
```

```python
#countplot function to count how many times a values occur in variable

sns.countplot(x=variable)
plt.show()
```

##### Seaborn and Pandas DF

```python
import pandas as pd
df = pd.read_csv("filename.csv")
```

```python
sns.countplot(x="column1", y="column2", data=df,
             hue="column3") 
            #distinguish datapoints by a second variable & adds legend automatically
plt.show()
```

```python
#you can also create a key of hue colors

hue_colors = {"value1":"black",
              "value2":"red"}

sns.countplot(x="column1", y="column2", data=df,
             hue="column3",
             palette=hue_colors) 
```
