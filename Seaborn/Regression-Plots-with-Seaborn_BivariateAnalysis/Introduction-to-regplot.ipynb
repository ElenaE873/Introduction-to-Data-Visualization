#### regplot vs. lmplot

- They are absolutely built with the same/similar arguments but lmplot() is much stronger.
- in lmplots the use of **hue** and **col** is really important for faceting
    - hue: creates a regression line based on the column designated and the values contained in it
    - col: creates subplots based on the values in the column specified
    - row(not displayed here): displays plots across multiple rows instead of vertically
