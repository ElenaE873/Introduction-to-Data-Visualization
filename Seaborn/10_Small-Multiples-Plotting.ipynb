Requires the data to be in tidy format: 1 observation per row and features must be in columns.

#### FacetGrid

- it provides an outlook for how the data is distributed across columns, rows and hue.

```python
#1 - Create a facetgrid
g = sns.FacetGrid(df, col="Column1")

#2 - map the boxplot to the facetgrid
g.map(sns.boxplot, 'Column2', 
     order = [1,2,3,4]) #<<order of values of the facetgrid column categ.

#also supports Scatter or Regression plots by switching plt.boxplot to
#plt.scatter
```

Example

#### Factorplot()

- faster/shortcut to creating a FacetGrid

```python
sns.factorplot(x='Column2', data=df,
              col='Column1', kind='box')
```

This accomplishes the exact thing above.

#### lmplot
- also provides a shortcut to plotting regression and scatterplots on FacetGrid

```python
sns.lmplot(data=df, 
           x="Column2", 
           y="Column3", 
           col="Column1", 
           fit_reg=False #eliminates the regression line
          )  

#also filter the rows by another column
sns.lmplot(data=df, 
           x="Column2", 
           y="Column3", 
           col="Column1", 
           row = "another_column" #filter column
          )
```

#### PairGrid
- show the relationship between data elements
- with this one the col/row parameters are not defined
- instead the variables are defined which is how the result gets organized

```python
g = sns.PairGrid(df, vars=["Column1", "Column2"])
g = g.map(plt.scatter)
```

Example

##### Customizing the Diagonals

```python
g=g.map_diag(plt.hist)
g=g.map_offdiag(plt.scatter)
```

Example

#### PairPlot

```python
g = sns.pairplot(df, vars=["Column1", "Column2"],
                kind='reg',
                diag_kind='hist')
```

#### JointGrid

```python
g.sns.JointGrid(data=df, x="Column1", y="Column2")
g.plot(sns.regplot, sns.displot)
```
